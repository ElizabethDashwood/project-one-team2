```python
# Dependencies
import requests
import json
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(color_codes=True)
pd.set_option('display.max_columns', None)
```

```python
# Read CSV file.
df = pd.read_csv('Melbourne_housing.csv')
df.head()
```

```python
# Check the number of unique values from all columns in the dataframe
df.nunique()
```

```python
# remove column with high unique values
df = df.drop(['Suburb', 'Address', 'SellerG'], axis=1)
df.head()
```

```python
# Extract the year from the Date column
df['Year'] = pd.DatetimeIndex(df['Date']).year
df.head()
```

```python
# Remove the rows with missing values
df = df.dropna()
df.head()

# Save the dataframe to a csv file
df.to_csv('melbourne_housing_cleaned.csv', index=False)
```

```python
# convert the date column to integer
df['Year'] = df['Year'].astype(int)
df.dtypes
```

```python
df['CouncilArea'].unique()
```

```python
df['CouncilArea'].unique()

def segment_council_area(council_area):
    if council_area in ['Yarra', 'Port Phillip', 'Stonnington', 'Boroondara']:
        return 'Inner Melbourne'
    elif council_area in ['Moonee Valley', 'Darebin', 'Hobsons Bay', 'Maribyrnong', 'Bayside', 'Moreland', 'Melbourne']:
        return 'Middle Melbourne'
    elif council_area in ['Monash', 'Glen Eira', 'Whitehorse', 'Banyule', 'Kingston']:
        return 'Outer Melbourne'
    else:
        return 'Other'

df['Segment'] = df['CouncilArea'].apply(segment_council_area)
df.head()
```

```python
sns.barplot(x='CouncilArea', y='Price', data=df)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 8))
sns.barplot(x='Price', y='CouncilArea', data=df)
plt.show()
```
